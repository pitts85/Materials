```python
import numpy as np
import pylab as plt

#Problem 6.2a 
def sprime6a(**kwargs):
    """
    For a cubic material, matrix compliances s=[s11,s12,s44],f
    find the stiffnesses in the sprime direction (on crystal axes), in plane with normal n,
    single in-plane direction u, angle theta CCW from u (about plane normal)
    s = np.array([15,0,-6.3,13.3])
    n =np.array([1,1,0])
    u = np.array([1,-1,0])
    th = 0.
    """
    opt={
    's':np.array([15,13.3,-6.3]), #TPa^-1: s11,s44,s12
    'n':np.array([1,0,0]),
    'u':np.array([0,1,0]),
    'th': 0.
    }
    
    opt.update(**kwargs)
    
    th=opt['th']; cos=np.cos; sin=np.sin
    
    #rotation_matrix = np.array([[cos(th),sin(th),0],[-sin(th),cos(th),0],[0,0,1.]])
    
    nn=opt['n']/np.linalg.norm(opt['n'])
    
    un0=opt['u']/np.linalg.norm(opt['u'])
    vn0=np.cross(nn,un0)
    
    un = cos(th)*un0+sin(th)*vn0
    vn =np.cross(nn,un)
    
    old_coords=np.array([[1.,0.,0.],[0,1.,0],[0,0,1.]])
    new_coords=np.array([un,vn,nn])
    
    a= np.dot(new_coords,old_coords)
    
    s=np.zeros((3,3,3,3))
    
    s[0,0,0,0]=opt['s'][0]
    s[1,1,1,1]=opt['s'][0]
    s[2,2,2,2]=opt['s'][0]
    
    s[0,0,1,1]=opt['s'][2]
    s[1,1,2,2]=opt['s'][2]
    s[2,2,0,0]=opt['s'][2]
    
    s[1,1,0,0]=opt['s'][2]
    s[2,2,1,1]=opt['s'][2]
    s[0,0,2,2]=opt['s'][2]
    
    s[0,1,0,1]=opt['s'][1]/4.
    s[1,2,1,2]=opt['s'][1]/4.
    s[2,0,2,0]=opt['s'][1]/4.
    
    s[1,0,1,0]=opt['s'][1]/4.
    s[2,1,2,1]=opt['s'][1]/4.
    s[0,2,0,2]=opt['s'][1]/4.
    
    s[0,1,1,0]=opt['s'][1]/4.
    s[1,2,2,1]=opt['s'][1]/4.
    s[2,0,0,2]=opt['s'][1]/4.
        
    s[1,0,0,1]=opt['s'][1]/4.
    s[2,1,1,2]=opt['s'][1]/4.
    s[0,2,2,0]=opt['s'][1]/4.
    
    retval = np.einsum('im,jn,ko,lp,mnop',a,a,a,a,s)
    
    return retval
    
#Problem 6.2b
def sprime6b(**kwargs):
    """
    For a cubic material, matrix compliances s=[s11,s12,s44],f
    find the stiffnesses in the sprime direction (on crystal axes), in plane with normal n,
    single in-plane direction u, angle theta CCW from u (about plane normal)
    s = np.array([15,0,-6.3,13.3])
    n =np.array([1,1,0])
    u = np.array([1,-1,0])
    th = 0.
    """
    opt={
    's':np.array([15,13.3,-6.3]), #TPa^-1: s11,s44,s12
    'n':np.array([1,1,1]),
    'u':np.array([1,1,0]),
    'th': 0.
    }
    
    opt.update(**kwargs)
    
    th=opt['th']; cos=np.cos; sin=np.sin
    
    #rotation_matrix = np.array([[cos(th),sin(th),0],[-sin(th),cos(th),0],[0,0,1.]])
    
    nn=opt['n']/np.linalg.norm(opt['n'])
    
    un0=opt['u']/np.linalg.norm(opt['u'])
    vn0=np.cross(nn,un0)
    
    un = cos(th)*un0+sin(th)*vn0
    vn =np.cross(nn,un)
    
    old_coords=np.array([[1.,0.,0.],[0,1.,0],[0,0,1.]])
    new_coords=np.array([un,vn,nn])
    
    a= np.dot(new_coords,old_coords)
    
    s=np.zeros((3,3,3,3))
    
    s[0,0,0,0]=opt['s'][0]
    s[1,1,1,1]=opt['s'][0]
    s[2,2,2,2]=opt['s'][0]
    
    s[0,0,1,1]=opt['s'][2]
    s[1,1,2,2]=opt['s'][2]
    s[2,2,0,0]=opt['s'][2]
    
    s[1,1,0,0]=opt['s'][2]
    s[2,2,1,1]=opt['s'][2]
    s[0,0,2,2]=opt['s'][2]
    
    s[0,1,0,1]=opt['s'][1]/4.
    s[1,2,1,2]=opt['s'][1]/4.
    s[2,0,2,0]=opt['s'][1]/4.
    
    s[1,0,1,0]=opt['s'][1]/4.
    s[2,1,2,1]=opt['s'][1]/4.
    s[0,2,0,2]=opt['s'][1]/4.
    
    s[0,1,1,0]=opt['s'][1]/4.
    s[1,2,2,1]=opt['s'][1]/4.
    s[2,0,0,2]=opt['s'][1]/4.
        
    s[1,0,0,1]=opt['s'][1]/4.
    s[2,1,1,2]=opt['s'][1]/4.
    s[0,2,2,0]=opt['s'][1]/4.
    
    retval = np.einsum('im,jn,ko,lp,mnop',a,a,a,a,s)
    
    return retval
    

def fig6a(**kwargs):
    thv=np.linspace(0,2.*np.pi,360)
    sv=np.array([sprime6a(th=thi,**kwargs)[0,0,0,0] for thi in thv])
    plt.close()
    plt.plot(thv,sv)
    plt.ylabel('Young\'s Modulus, $S^{\prime}_{1111}$')
    plt.xlabel('Angle (Radians)')
    plt.savefig(r"/Users/andrewpitts/Documents/Columbia/MSAE E4215 Mechanical Behavior of Materials/HW 6/fig6-a.png")
    
def fig6b(**kwargs):
    thv=np.linspace(0,2.*np.pi,360)
    sv=np.array([sprime6b(th=thi,**kwargs)[0,0,0,0] for thi in thv])
    plt.close()
    plt.plot(thv,sv)
    plt.ylabel('Young\'s Modulus, $S^{\prime}_{1111}$')
    plt.xlabel('Angle (Radians)')
    plt.savefig(r"/Users/andrewpitts/Documents/Columbia/MSAE E4215 Mechanical Behavior of Materials/HW 6/fig6-b.png")
    
fig6a()
fig6b()
```
